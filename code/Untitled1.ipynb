```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


x = dataset.loc[:, names].values



x = StandardScaler().fit_transform(x) # normalizing the features

x.shape
np.mean(x),np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]

normalised_COVID = pd.DataFrame(x,columns=feat_cols)
normalised_COVID.tail()
```

```python
from sklearn.decomposition import PCA
pca_COVID = PCA(n_components=7)
principalComponents_COVID = pca_COVID.fit_transform(x)
#pca_COVID.fit(normalised_COVID)







principal_COVID_Df = pd.DataFrame(data = principalComponents_COVID, columns = ['principal component 1', 'principal component 2','principal component 3','principal component 4','principal component 5','principal component 6','principal component 7'])
principal_COVID_Df.tail()

principal_COVID_Df
```

```python
x_pca = pca_COVID.transform(normalised_COVID)
normalised_COVID.shape
```

```python
print('Explained variation per principal component: {}'.format(pca_COVID.explained_variance_ratio_))
```
